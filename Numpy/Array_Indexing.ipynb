## 1. Numpy indexing and slicing of vectors

```python
arr = np.arange(11,21)
```

```python
arr
```

```python
arr[9]
```

```python
arr[1:5]
```

```python
arr_slice = arr[3::2]
arr_slice
```

```python
arr_slice
```

## 2. Numpy Indexing and slicing of matrix

```python
arr = np.arange(1,31).reshape(6,5)
arr
```

```python
arr[5,4]
```

```python
slice = arr[0:2,1:3]
slice
```

```python
arr[3:6,3:6]
```

```python
arr[:,2]
```

## 3. Boolean Indexing

```python
arr = np.arange(11,21)
arr
```

```python
bool_index = arr % 2 == 0
bool_index
```

```python
arr = arr[bool_index]
arr
```
